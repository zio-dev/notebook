```python
from torchvision import datasets, transforms

# トランスフォームの設定（MNISTの画像をTensorに変換）
transform = transforms.Compose([
    transforms.ToTensor(),  # 画像データをTensorに変換
    transforms.Normalize((0.5,), (0.5,))  # 正規化（画像のピクセル値を-1から1の範囲に変換）
])

# MNISTデータセットをダウンロードし、指定したディレクトリに保存
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

print("トレーニングデータセットのサイズ:", len(train_dataset))
print("テストデータセットのサイズ:", len(test_dataset))
```

```python
import matplotlib.pyplot as plt
import torch
import torchvision
import numpy as np

# DataLoaderを使用して、データを扱いやすくする
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)

# データローダーから一つのバッチを取得し、画像とラベルを取り出す
data_iter = iter(train_loader)
images, labels = next(data_iter)

# 画像を表示する関数
def imshow(img):
    img = img / 2 + 0.5  # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# 最初の画像を表示（複数の画像をグリッド表示）
imshow(torchvision.utils.make_grid(images))

# 対応するラベルを表示（最初の画像のラベル）
print('Label:', labels[0].item())
```

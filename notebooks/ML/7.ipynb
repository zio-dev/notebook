```python
# 解析する文
text = """
吾輩は猫である。名前はまだない。どこで生れたか頓（とん）と見当がつかぬ。
何でも薄暗いじめじめした所でニャーニャー泣いていた事だけは記憶している。吾輩はここで始めて人間というものを見た。
しかもあとで聞くとそれは書生という人間中で一番獰悪（どうあく）な種族であったそうだ。この書生というのは時々我々を捕（つかま）えて煮て食うという話である。
"""
```

```python
import spacy

# GiNZAモデルをロード
nlp = spacy.load('ja_ginza')
```

```python
# テキストを処理
doc = nlp(text)

# トークンごとにテキスト、品詞タグ、詳細なタグを出力
for token in doc:
    print(token.text, token.pos_, token.tag_)

# 固有表現を抽出して表示
for ent in doc.ents:
    print(ent.text, ent.label_)

# 依存関係の解析結果を表示
for token in doc:
    print(token.text, token.dep_, token.head.text)
```

```python
text = 'ラインアップ SUV LANDCRUISER“250” 進化した走行性能と機能性を高めたデザイン。さまざまな用途に応えるランドクルーザーの中核モデル。 オフローダーとしての基本性能を向上 パワフルな走りを実現 機能性に磨きをかけたデザイン GR/GR SPORT GRヤリス モータースポーツを起点としたもっといいクルマづくり。その終わりなき挑戦を続け、進化し続ける。 勝つためのデザイン ドライバーファーストの探求 モータースポーツで培った技術を楽しさへ SUV LANDCRUISER“70” ランドクルーザーの原点が、国内復活。荒れた路面や、過酷な状況も乗り越えて行ける、本格オフローダー。 タフな動力性能と低燃費を堂々両立 強靱な、伝統のラダーフレームを継承 シンプルだからこそ、使いやすく、心強い セダン CROWN FCEVという新たな選択肢を加え、ニューフォーマルに進化したクラウン。 威厳と風格を備えた、美しい佇まい。 室内にひろがるおもてなし空間。 優雅で心地良い走り。 SUV CROWN"SPORT" エモーショナルで創造的な雰囲気とあらゆる人を笑顔にする走りで、あなたの心を弾ませるスポーツSUV。 目を奪う艶やかなエクステリアデザイン 上質、俊敏。パワフルなPHEVモデルと、 爽快なハイブリッドモデル。'
texts = text.split(' ')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TfidfVectorizerの初期化
vectorizer = TfidfVectorizer()

# テキストをベクトル化
tfidf_matrix = vectorizer.fit_transform(texts)

# 特徴名（単語）の取得
feature_names = vectorizer.get_feature_names_out()

# 各ドキュメントのTF-IDF値を出力
print("TF-IDF Values:")
for i in range(len(texts)):
    print(f"Document {i+1}:")
    # TF-IDF値が格納された行列からドキュメントを取得し、スコアでソート
    doc = tfidf_matrix[i].toarray().flatten()
    sorted_indices = doc.argsort()[::-1]  # スコアが高い順に並び替え
    for idx in sorted_indices:
        if doc[idx] > 0:  # スコアが0より大きい単語のみ出力
            print(f"  {feature_names[idx]}: {doc[idx]:.4f}")
```

```python
from sudachipy import dictionary
from sudachipy.tokenizer import Tokenizer

# Sudachiのトークナイザー設定
class SudachiTokenizer:
    def __init__(self):
        self.tokenizer = dictionary.Dictionary().create()

    def __call__(self, text):
        return [m.surface() for m in self.tokenizer.tokenize(text, Tokenizer.SplitMode.C)]
```

```python
# サンプルテキスト
texts = [
    "猫は夜中に活動的です。",
    "犬は散歩が大好きです。",
    "猫と犬どちらもペットとして人気があります。"
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sudachiトークナイザーのインスタンス作成
tokenizer = SudachiTokenizer()

# TfidfVectorizerの初期化、Sudachiトークナイザーを指定
vectorizer = TfidfVectorizer(tokenizer=tokenizer, token_pattern=None)

# ドキュメントをベクトル化
tfidf_matrix = vectorizer.fit_transform(texts)

# 特徴名（単語）の取得
feature_names = vectorizer.get_feature_names_out()

# 各ドキュメントのTF-IDF値を出力
print("TF-IDF Values:")
for i in range(len(texts)):
    print(f"Document {i+1}:")
    doc = tfidf_matrix[i].toarray().flatten()
    sorted_indices = doc.argsort()[::-1]  # スコアが高い順に並び替え
    for idx in sorted_indices:
        if doc[idx] > 0:  # スコアが0より大きい単語のみ出力
            print(f"  {feature_names[idx]}: {doc[idx]:.4f}")
```

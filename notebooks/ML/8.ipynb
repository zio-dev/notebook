```python
import numpy as np
import matplotlib.pyplot as plt

# 乱数のシードを設定
np.random.seed(0)

# 家の面積データを生成 (平均120平方メートル、標準偏差30平方メートル)
X = 30 * np.random.randn(100) + 120

# 残差項を生成 (この場合、価格の変動を表す小さなノイズ)
res = 20 * np.random.randn(100)  # 価格の変動を20万円の標準偏差で設定

# 家の価格を計算 (基本価格を3000万円とし、面積1平方メートルあたりの価格を10万円とする)
y = 3000 + 10 * X + res

# データのプロット
plt.scatter(X, y)
plt.title('House Price vs. Size')
plt.xlabel('Size in square meters')
plt.ylabel('Price in ten thousand yen')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# データを訓練用とテスト用に分割
X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1, 1), y, test_size=0.2, random_state=0)

# 線形回帰モデルの作成
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# テストデータに対する予測
y_pred = model.predict(X_test)

# モデルの評価
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean squared error: {:.2f}".format(mse))
print("Coefficient of determination (R^2): {:.2f}".format(r2))

# テストデータと予測結果のプロット
plt.scatter(X_test, y_test, color='black', label='Actual Price')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Predicted Price')
plt.title('Test Data and Predictions')
plt.xlabel('Size in square meters')
plt.ylabel('Price in ten thousand yen')
plt.legend()
plt.show()
```

### 関数定義

```python
import requests
from bs4 import BeautifulSoup


def crawl_html_text(url:str)->str:
    response = requests.get(url)
    response.encoding = 'utf-8'
    return response.text

def remove_tag(source:str)->str:
    soup = BeautifulSoup(source, 'html.parser')

    remove_target = [
        'a',
        'button',
        'img',
        'link',
        'meta',
        'script',
        'style',
        'svg',
    ]

    for tag in remove_target:
        for element in soup.find_all(tag):
            element.decompose()

    for tag in soup.findAll(True):
        if 'style' in tag.attrs:
            del tag.attrs['style']

        if 'class' in tag.attrs:
            del tag.attrs['class']

    return str(soup)
```

### ソースの取得

```python
target_url = 'https://toyota.jp/index.html'
html_text = crawl_html_text(target_url)
remove_tag_source = remove_tag(html_text)
print(remove_tag_source)
```

### 文字の抜き出し

```python
soup = BeautifulSoup(remove_tag_source, 'html.parser')
print(' '.join(soup.get_text().split()))
```
